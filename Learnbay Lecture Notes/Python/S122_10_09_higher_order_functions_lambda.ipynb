```python
# filter
# filter(funct, seq): function should return a boolean
# if function returns true for element, it's kept in result otherwise not
```

```python
a = [11,22,33,44,55,55,66]

def is_even(x):
    return x%2 == 0

r = filter(is_even, a) 
print(r)
print(list(r))
```

```python
a = ["Gaurav", "Abhishek", "Krishna", "Apporva"]

def equal_first_last(s):
    return s[0] == s[-1]
 
r = filter(equal_first_last, map(str.upper, a))
print(list(r))
```

```python
def f(funct, seq): # filter equivalent.. returns list instead of iterator
    a = []
    for v in seq:
        if funct(v):
            a.append(v)
    return a

a = ["Gaurav", "Abhishek", "Krishna", "Apporva"]

def equal_first_last(s):
    return s[0] == s[-1]
 
r = f(equal_first_last, map(str.upper, a))
print(r)
```

```python
# reduce
```

```python
def f(funct, seq):
    it = iter(seq)
    x = next(it)

    for y in it:
        x = funct(x,y)
    return x

def add(x,y):
    return x + y

a = [10,20,30,40]
r =  f(add, a)
print(r)


# x 10 10  30  60 100
# y    20  20  40 
```

```python
from functools import reduce
def add(x,y):
    return x + y

a = [10,20,30,40]
r = reduce(add, a)
print(r)
```

```python
# function passed to reduce should take exactly 2 arguments
from functools import reduce
def add(x,y,z):
    return x + y + z

a = [10,20,30,40]
r = reduce(add, a)
print(r)
```

```python
from functools import reduce

def my_min(x,y):
    return x if x < y else y

a = [110,20,30,40]
r = reduce(my_min, a)
print(r)
print(min(a))
```

```python
# lambda
- anonymous function: nameless
- they are only one liner
- can take args
- cannot contain a return
- they should contain an expression whose result is automatically returned
```

```python
lambda <args> : <expression>
```

```python
x = lambda : 10 # zero arg lambda
print(x())
print(x())
```

```python
x = lambda x: x**2 # 1 arg lambda
print(x(10))
print(x(2))
```

```python
# sort on basis of last alphabet
a = ["Gaurav", "Abhishek", "Krishna", "apporva"]

a.sort(key=lambda s:s[-1])
print(a)
```

```python
from functools import reduce

a = [10,20,30,40]

# using function
def add(x,y):
    return x + y

r = reduce(add, a)
print(r)

# using lambda
r = reduce(lambda x,y: x + y, a)
print(r)
```
