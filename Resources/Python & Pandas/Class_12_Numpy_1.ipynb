# Arithmetic Operation on Numpy array

```python
import numpy as np
```

```python
arr1 = np.array([10,11,12,13,14,15])
arr2 = np.array([20,21,22,23,24,25])
```

```python
print(arr1 + arr2)
```

```python
print(arr2 - arr1)
```

```python
print(arr2/arr1)
```

```python
print(arr2//arr1)
```

```python
arr2 ** arr1
```

```python
arr2 % arr1
```

```python
arr2 > 23
```

```python
arr2 < 23
```

```python
arr2 * arr1
```

```python
# dot product

# [[10,20],        [[2,3],
# [30,40]]         [4,5]]

# [[100, 130],
# [220, 290]]

arr1 = np.array([[10,20],[30,40]])
arr2 = np.array([[2,3],[4,5]])
print(arr1 * arr2)
print()
print(arr1.dot(arr2))
```
